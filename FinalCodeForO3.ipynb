## Autoencoder(feature Extraction) + CNN (Classification)

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, f1_score
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout, Input
from tensorflow.keras.utils import to_categorical
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create output directory
output_folder = "Autoencoder_CNN_Outputs"
os.makedirs(output_folder, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)
y = data['Condition']
gene_names = X.columns  # Save gene names for reference if needed

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

### Autoencoder for Feature Extraction
def build_autoencoder(input_dim, encoding_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(encoding_dim, activation='relu')(input_layer)
    decoded = Dense(input_dim, activation='sigmoid')(encoded)
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    encoder = Model(input_layer, encoded)
    return autoencoder, encoder

def apply_autoencoder(X_train, X_test, encoding_dim=500):
    input_dim = X_train.shape[1]
    autoencoder, encoder = build_autoencoder(input_dim, encoding_dim)
    autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, shuffle=True, validation_data=(X_test, X_test), verbose=1)
    X_train_encoded = encoder.predict(X_train)
    X_test_encoded = encoder.predict(X_test)
    return X_train_encoded, X_test_encoded

X_train_encoded, X_test_encoded = apply_autoencoder(X_train_scaled, X_test_scaled, encoding_dim=500)

# Save encoded features identified by the autoencoder with new feature names
encoded_feature_names = [f'Encoded_Feature_{i+1}' for i in range(X_train_encoded.shape[1])]
encoded_features_df = pd.DataFrame(X_train_encoded, columns=encoded_feature_names)

# Calculate mean importance as a reference for sorting
encoded_features_mean_importance = encoded_features_df.mean(axis=0).sort_values(ascending=False)

# Create a DataFrame with encoded feature names and corresponding importance scores
encoded_features_with_names = pd.DataFrame(encoded_features_mean_importance, columns=['Importance'])
encoded_features_with_names.to_csv(os.path.join(output_folder, "autoencoder_CNN_encoded_features_with_importance.csv"), index=True)

# Reshape data for CNN
X_train_encoded_reshaped = X_train_encoded.reshape(X_train_encoded.shape[0], X_train_encoded.shape[1], 1)
X_test_encoded_reshaped = X_test_encoded.reshape(X_test_encoded.shape[0], X_test_encoded.shape[1], 1)

### CNN Classifier
def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(64, 3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.3))
    model.add(Conv1D(32, 3, activation='relu'))
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(3, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

cnn_model = build_cnn((X_train_encoded_reshaped.shape[1], 1))

# Train the CNN model and save training history
start_time = time.time()
history = cnn_model.fit(X_train_encoded_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)
training_time = time.time() - start_time

# Plot and save training history
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.savefig(os.path.join(output_folder, "autoencoder_CNN_training_history.png"), dpi=300)
plt.close()

# Evaluate the model
loss, accuracy = cnn_model.evaluate(X_test_encoded_reshaped, y_test)

# Predictions and Metrics
y_pred = cnn_model.predict(X_test_encoded_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

classification_rep = classification_report(y_test_classes, y_pred_classes, output_dict=True, target_names=label_encoder.classes_)
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save model metrics
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC', 'Training Time'],
    'Score': [
        accuracy,
        classification_rep['weighted avg']['precision'],
        classification_rep['weighted avg']['recall'],
        classification_rep['weighted avg']['f1-score'],
        roc_auc,
        training_time
    ]
})
metrics_df.to_csv(os.path.join(output_folder, "autoencoder_CNN_model_metrics.csv"), index=False)

# Plot Confusion Matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_folder, "autoencoder_CNN_confusion_matrix.png"), dpi=300)
plt.close()

# Plot ROC Curve
plt.figure(figsize=(8, 6))
fpr = {}
tpr = {}
for i, class_name in enumerate(label_encoder.classes_):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr[i], tpr[i], label=f'{class_name} (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.savefig(os.path.join(output_folder, "autoencoder_CNN_ROC_curve.png"), dpi=300)
plt.close()

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    precision, recall, _ = precision_recall_curve(y_test[:, i], y_pred[:, i])
    plt.plot(recall, precision, label=f'{class_name}')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.savefig(os.path.join(output_folder, "autoencoder_CNN_precision_recall_curve.png"), dpi=300)
plt.close()

# Plot F1-Score Bar Chart
f1_scores = [classification_rep[class_name]['f1-score'] for class_name in label_encoder.classes_]
plt.figure(figsize=(8, 6))
plt.bar(label_encoder.classes_, f1_scores, color='skyblue')
plt.xlabel("Class")
plt.ylabel("F1 Score")
plt.title("F1 Score by Class")
plt.savefig(os.path.join(output_folder, "autoencoder_CNN_f1_score.png"), dpi=300)
plt.close()

print("All results saved successfully in Autoencoder_CNN_Outputs folder.")
```

## PCA + CNN

```python
import os
import tensorflow as tf
import numpy as np
import random
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import (classification_report, confusion_matrix, roc_curve, 
                             auc, precision_recall_fscore_support)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.optimizers import Nadam
from tensorflow.keras.utils import to_categorical
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Create output folder
output_folder = "PCA_CNN"
os.makedirs(output_folder, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)
y = data['Condition']

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply PCA
def apply_pca(X_train, X_test, feature_names, output_folder, n_components=1000, top_features=1000):
    # Perform PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_
    
    # Get the top features by explained variance
    top_indices = explained_variance.argsort()[::-1][:top_features]
    top_variance_scores = explained_variance[top_indices]
    top_feature_names = feature_names[top_indices]
    
    # Save top features in specified format
    top_features_df = pd.DataFrame({
        'Feature Name': top_feature_names,
        'Score': top_variance_scores
    })
    top_features_df.to_csv(os.path.join(output_folder, "top_pca_features.csv"), index=False)
    
    # Return transformed data and PCA object
    return X_train_pca, X_test_pca, pca

# Pass feature names (X_train.columns) to apply_pca function
X_train_pca, X_test_pca, pca = apply_pca(X_train_scaled, X_test_scaled, feature_names=X.columns, output_folder=output_folder, n_components=1000)

# Reshape data for CNN
X_train_pca_reshaped = X_train_pca.reshape(X_train_pca.shape[0], X_train_pca.shape[1], 1)
X_test_pca_reshaped = X_test_pca.reshape(X_test_pca.shape[0], X_test_pca.shape[1], 1)

# Build CNN model
def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(64, kernel_size=5, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.3))
    model.add(Conv1D(64, kernel_size=5, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(len(label_encoder.classes_), activation='softmax'))
    
    model.compile(optimizer=Nadam(learning_rate=0.007), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train CNN model
cnn_model = build_cnn((X_train_pca_reshaped.shape[1], 1))
history = cnn_model.fit(X_train_pca_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)

# Evaluate model
loss, accuracy = cnn_model.evaluate(X_test_pca_reshaped, y_test)

# Model predictions and metrics
y_pred = cnn_model.predict(X_test_pca_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Save classification report
classification_report_dict = classification_report(y_test_classes, y_pred_classes, 
                                                   target_names=label_encoder.classes_, 
                                                   output_dict=True)
classification_report_df = pd.DataFrame(classification_report_dict).transpose()
classification_report_df.to_csv(os.path.join(output_folder, "classification_report.csv"))

# Save confusion matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
conf_matrix_df.to_csv(os.path.join(output_folder, "confusion_matrix.csv"))

# Calculate precision, recall, F1-score, AUC-ROC
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average='weighted')
fpr, tpr, thresholds = roc_curve(y_test_classes, y_pred[:, 1], pos_label=1)  # Adjust for positive class
roc_auc = auc(fpr, tpr)

# Save model metrics
model_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1_score,
    'AUC-ROC': roc_auc,
    'Training Time (epochs)': history.epoch[-1]
}
model_metrics_df = pd.DataFrame(model_metrics, index=[0])
model_metrics_df.to_csv(os.path.join(output_folder, "model_metrics.csv"), index=False)

# Save training history
history_df = pd.DataFrame(history.history)
history_df.to_csv(os.path.join(output_folder, "training_history.csv"), index=False)

# Save sorted PCA features
pca_features_sorted = pd.DataFrame({
    'PCA_Feature': [f'PCA_Feature_{i+1}' for i in range(len(pca.explained_variance_ratio_))],
    'Explained_Variance': pca.explained_variance_ratio_
}).sort_values(by='Explained_Variance', ascending=False)
pca_features_sorted.to_csv(os.path.join(output_folder, "sorted_pca_features.csv"), index=False)

# Plot Confusion Matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_folder, "confusion_matrix_plot.png"), dpi=300)
plt.close()

# Plot ROC Curve
plt.figure(figsize=(10, 7))
RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name="PCA_CNN").plot()
plt.title("ROC Curve")
plt.savefig(os.path.join(output_folder, "roc_curve_plot.png"), dpi=300)
plt.close()

# Plot Training and Validation Accuracy and Loss
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_folder, "training_validation_plot.png"), dpi=300)
plt.close()

# Plot Precision, Recall, F1-Score Bar Chart
metrics = ['Precision', 'Recall', 'F1-Score']
scores = [precision, recall, f1_score]

plt.figure(figsize=(8, 5))
sns.barplot(x=metrics, y=scores)
plt.title('Precision, Recall, and F1-Score')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.savefig(os.path.join(output_folder, "precision_recall_f1_plot.png"), dpi=300)
plt.close()

print("All results and plots saved successfully in PCA_CNN folder!")

# for auc-roc

# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "PCA_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "PCA_CNN_ROC_curve.png"), dpi=300)
plt.close()

```

## MI + CNN

```python
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.optimizers import Nadam
from tensorflow.keras.utils import to_categorical
import seaborn as sns
import matplotlib.pyplot as plt

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Set up output directory
output_folder = "MI_CNN_Outputs"
os.makedirs(output_folder, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)  # Drop non-feature columns
y = data['Condition']  # Target variable: Condition

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Feature Extraction using Mutual Information
def apply_mi(X_train, y_train, X_test, feature_names, n_features=1000):
    mi_scores = mutual_info_classif(X_train, y_train.argmax(axis=1))
    top_n_features = np.argsort(mi_scores)[-n_features:]
    
    # Store the top features and their MI scores
    top_features_df = pd.DataFrame({
        'Feature Name': feature_names[top_n_features],
        'MI Score': mi_scores[top_n_features]
    })
    top_features_df.to_csv(os.path.join(output_folder, "top_mi_features.csv"), index=False)

    X_train_mi = X_train[:, top_n_features]
    X_test_mi = X_test[:, top_n_features]
    
    print(f"MI applied: Selected top {n_features} features.")
    return X_train_mi, X_test_mi

# Apply MI feature selection
X_train_mi, X_test_mi = apply_mi(X_train_scaled, y_train, X_test_scaled, feature_names=X.columns, n_features=1000)

# Reshape data for CNN
X_train_mi_reshaped = X_train_mi.reshape(X_train_mi.shape[0], X_train_mi.shape[1], 1)
X_test_mi_reshaped = X_test_mi.reshape(X_test_mi.shape[0], X_test_mi.shape[1], 1)

# CNN model
def build_cnn(input_shape):
    model = Sequential([
        Conv1D(64, kernel_size=5, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        Dropout(0.5),
        Conv1D(32, kernel_size=3, activation='relu'),
        MaxPooling1D(pool_size=2),
        Dropout(0.5),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.5),
        Dense(3, activation='softmax')
    ])
    model.compile(optimizer=Nadam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the model
cnn_model = build_cnn((X_train_mi_reshaped.shape[1], 1))
history = cnn_model.fit(X_train_mi_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)

# Model evaluation
loss, accuracy = cnn_model.evaluate(X_test_mi_reshaped, y_test)
print(f"Test Accuracy: {accuracy:.2f}")

# Predictions and metrics
y_pred = cnn_model.predict(X_test_mi_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Classification report
classification_report_dict = classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_, output_dict=True)
classification_report_df = pd.DataFrame(classification_report_dict).transpose()
classification_report_df.to_csv(os.path.join(output_folder, "classification_report.csv"))

# Confusion matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
conf_matrix_df.to_csv(os.path.join(output_folder, "confusion_matrix.csv"))

# Save model metrics
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average='weighted')
model_metrics = {
    'Accuracy': accuracy,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1_score,
}
pd.DataFrame(model_metrics, index=[0]).to_csv(os.path.join(output_folder, "model_metrics.csv"), index=False)

# Save training history
history_df = pd.DataFrame(history.history)
history_df.to_csv(os.path.join(output_folder, "training_history.csv"), index=False)

# Plot and save confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_folder, "confusion_matrix_plot.png"), dpi=300)
plt.close()

# Plot training and validation curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_folder, "training_validation_plot.png"), dpi=300)
plt.close()

print("All outputs saved in MI_CNN_Outputs folder!")



# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "MI_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "MI_CNN_ROC_curve.png"), dpi=300)
plt.close()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import time
import os  # For creating directories
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, roc_curve,
    precision_recall_curve, auc, f1_score, accuracy_score, precision_score, recall_score
)
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout, Input
from tensorflow.keras.utils import to_categorical

# Create output directory
output_dir = 'Autoencoder_MI'
os.makedirs(output_dir, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Adjust file path as needed

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)
y = data['Condition']

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Autoencoder for Dimensionality Reduction
def build_autoencoder(input_dim, encoding_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(encoding_dim, activation='relu')(input_layer)
    decoded = Dense(input_dim, activation='sigmoid')(encoded)
    autoencoder = Model(input_layer, decoded)
    encoder = Model(input_layer, encoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder

def apply_autoencoder(X_train, X_test, encoding_dim=500):
    input_dim = X_train.shape[1]
    autoencoder, encoder = build_autoencoder(input_dim, encoding_dim)
    autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, shuffle=True, 
                    validation_data=(X_test, X_test), verbose=1)
    X_train_encoded = encoder.predict(X_train)
    X_test_encoded = encoder.predict(X_test)
    return X_train_encoded, X_test_encoded

def apply_mutual_info(X_train, X_test, y_train, num_features=500):
    mi_selector = SelectKBest(mutual_info_classif, k=num_features)
    X_train_mi = mi_selector.fit_transform(X_train, np.argmax(y_train, axis=1))
    X_test_mi = mi_selector.transform(X_test)
    selected_features = mi_selector.get_support(indices=True)
    selected_feature_names = X.columns[selected_features]
    important_features_df = pd.DataFrame({
        'Gene': selected_feature_names,
        'Score': mi_selector.scores_[selected_features]
    })
    important_features_df.to_csv(f'{output_dir}/CNN_important_features.csv', index=False)  # Save important features
    return X_train_mi, X_test_mi

X_train_encoded, X_test_encoded = apply_autoencoder(X_train_scaled, X_test_scaled, encoding_dim=1000)
X_train_hybrid, X_test_hybrid = apply_mutual_info(X_train_encoded, X_test_encoded, y_train, num_features=500)

# Reshape data for CNN
X_train_hybrid_reshaped = X_train_hybrid.reshape(X_train_hybrid.shape[0], X_train_hybrid.shape[1], 1)
X_test_hybrid_reshaped = X_test_hybrid.reshape(X_test_hybrid.shape[0], X_test_hybrid.shape[1], 1)

def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(64, 3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.3))
    model.add(Conv1D(32, 3, activation='relu'))
    model.add(MaxPooling1D(2))
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(3, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the CNN model
cnn_model = build_cnn((X_train_hybrid_reshaped.shape[1], 1))
start_time = time.time()
history = cnn_model.fit(X_train_hybrid_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)
training_time = time.time() - start_time

# Evaluate the model
loss, accuracy = cnn_model.evaluate(X_test_hybrid_reshaped, y_test)
y_pred = cnn_model.predict(X_test_hybrid_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Compute Metrics
precision = precision_score(y_test_classes, y_pred_classes, average='weighted')
recall = recall_score(y_test_classes, y_pred_classes, average='weighted')
f1 = f1_score(y_test_classes, y_pred_classes, average='weighted')
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save Model Metrics
metrics_df = pd.DataFrame({
    'Model': ['CNN'],
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall (Sensitivity)': [recall],
    'F1-Score': [f1],
    'AUC-ROC': [roc_auc],
    'Training Time (s)': [training_time]
})
metrics_df.to_csv(f'{output_dir}/CNN_model_metrics.csv', index=False)

# Plotting Functions
def plot_loss(history):
    plt.figure()
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig(f'{output_dir}/training_validation_loss.png', dpi=300)

def plot_accuracy(history):
    plt.figure()
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.savefig(f'{output_dir}/training_validation_accuracy.png', dpi=300)

def plot_roc(y_test, y_pred):
    fpr, tpr, _ = roc_curve(y_test.ravel(), y_pred.ravel())
    plt.figure()
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.savefig(f'{output_dir}/roc_curve.png', dpi=300)

def plot_precision_recall(y_test, y_pred):
    precision_vals, recall_vals, _ = precision_recall_curve(y_test.ravel(), y_pred.ravel())
    plt.figure()
    plt.plot(recall_vals, precision_vals, label='Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()
    plt.savefig(f'{output_dir}/precision_recall_curve.png', dpi=300)

def plot_confusion_matrix(y_test_classes, y_pred_classes):
    conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
    plt.figure()
    plt.matshow(conf_matrix, cmap='Blues')
    plt.colorbar()
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.savefig(f'{output_dir}/confusion_matrix.png', dpi=300)

# Generate and save plots
plot_loss(history)
plot_accuracy(history)
plot_roc(y_test, y_pred)
plot_precision_recall(y_test, y_pred)
plot_confusion_matrix(y_test_classes, y_pred_classes)

print("All results saved in the 'Autoencoder_MI' folder.")


# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "autoencoder+MI_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "autoencoder+MI_CNN_ROC_curve.png"), dpi=300)
plt.close()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create output folder
output_folder = "PCA_MI_CNN"
os.makedirs(output_folder, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)  # Drop non-feature columns
y = data['Condition']  # Target variable: Condition

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert class labels to numerical form (e.g., 0, 1, 2)
y_categorical = to_categorical(y_encoded)   # One-hot encode target for neural network

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape the data for CNN
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)  # (samples, features, 1)
X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

### Feature Extraction

# Method 1: PCA for Dimensionality Reduction
def apply_pca(X_train, X_test, n_components=1000):  # Increase to 500 or 1000
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    print(f"Explained variance by PCA: {sum(pca.explained_variance_ratio_):.2f}")
    return X_train_pca, X_test_pca

# Method 2: Mutual Information for Feature Selection
def apply_mutual_info(X_train, X_test, y_train, num_features=500):  # Increase to 500 or 1000
    mi_selector = SelectKBest(mutual_info_classif, k=num_features)
    X_train_mi = mi_selector.fit_transform(X_train, np.argmax(y_train, axis=1))  # Select top features based on Mutual Information
    X_test_mi = mi_selector.transform(X_test)
    selected_features = mi_selector.get_support(indices=True)
    selected_feature_names = X.columns[selected_features]  # Get the names of selected features
    print(f"Selected top {num_features} features based on Mutual Information")

    # Save selected important features to a CSV file
    important_features_df = pd.DataFrame(selected_feature_names, columns=['Important_Features'])
    important_features_df.to_csv(os.path.join(output_folder, 'important_features.csv'), index=False)
    print(f"Important features saved to 'important_features.csv'")
    
    return X_train_mi, X_test_mi

# Hybrid Feature Selection (PCA + Mutual Information)
X_train_pca, X_test_pca = apply_pca(X_train_scaled, X_test_scaled, n_components=1000)  # Increase to 500 or 1000
X_train_hybrid, X_test_hybrid = apply_mutual_info(X_train_pca, X_test_pca, y_train, num_features=500)  # Increase to 500 or 1000

# Reshape the data for CNN
X_train_hybrid_reshaped = X_train_hybrid.reshape(X_train_hybrid.shape[0], X_train_hybrid.shape[1], 1)
X_test_hybrid_reshaped = X_test_hybrid.reshape(X_test_hybrid.shape[0], X_test_hybrid.shape[1], 1)

### Convolutional Neural Network (CNN) Classifier

# Build the model
def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(len(label_encoder.classes_), activation='softmax'))  # Adjust for the number of classes
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the CNN model
cnn_model = build_cnn((X_train_hybrid_reshaped.shape[1], 1))

# Fit the model on training data
history = cnn_model.fit(X_train_hybrid_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)

# Evaluate the model on test data
loss, accuracy = cnn_model.evaluate(X_test_hybrid_reshaped, y_test)
print(f"Test Accuracy: {accuracy:.2f}")

### Model Evaluation

# Predict on the test set
y_pred = cnn_model.predict(X_test_hybrid_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Classification Report
classification_rep = classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_, output_dict=True)
classification_report_df = pd.DataFrame(classification_rep).transpose()
classification_report_df.to_csv(os.path.join(output_folder, "classification_report.csv"))

print("Classification Report:\n", classification_report_df)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
conf_matrix_df.to_csv(os.path.join(output_folder, "confusion_matrix.csv"))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_folder, "confusion_matrix_plot.png"), dpi=300)
plt.show()

### Plotting Training and Validation Curves

# Plotting the training and validation accuracy and loss
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_folder, "training_validation_plot.png"), dpi=300)
plt.show()

# Save training history
history_df = pd.DataFrame(history.history)
history_df.to_csv(os.path.join(output_folder, "training_history.csv"), index=False)

# Save model metrics
model_metrics = {
    'Loss': loss,
    'Accuracy': accuracy,
}
model_metrics_df = pd.DataFrame(model_metrics, index=[0])
model_metrics_df.to_csv(os.path.join(output_folder, "model_metrics.csv"), index=False)

print("All results and plots saved successfully in the PCA_MI_CNN folder!")

# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "PCA+MI_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "PCA+MI_CNN_ROC_curve.png"), dpi=300)
plt.close()

```

```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import Lasso
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
import random

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Create output folder
output_folder = "Lasso_CNN"
os.makedirs(output_folder, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)  # Drop non-feature columns
y = data['Condition']  # Target variable: Condition

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

### Feature Selection using Lasso
def apply_lasso(X_train, y_train, X_test, output_folder, alpha=0.01, n_features=1000):
    # Train a Lasso model
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train, y_train.argmax(axis=1))
    
    # Get the top n features
    coefs = lasso.coef_
    top_n_features = np.argsort(np.abs(coefs))[-n_features:]
    
    # Save selected features
    selected_features_df = pd.DataFrame({
        'Feature Index': top_n_features,
        'Coefficient': coefs[top_n_features]
    })
    selected_features_df.to_csv(os.path.join(output_folder, "top_lasso_features.csv"), index=False)
    
    X_train_lasso = X_train[:, top_n_features]
    X_test_lasso = X_test[:, top_n_features]
    print(f"Lasso applied: Selected top {n_features} features.")
    return X_train_lasso, X_test_lasso

# Apply Lasso
X_train_lasso, X_test_lasso = apply_lasso(X_train_scaled, y_train, X_test_scaled, output_folder=output_folder, alpha=0.01, n_features=1000)

# Reshape the data for CNN
X_train_lasso_reshaped = X_train_lasso.reshape(X_train_lasso.shape[0], X_train_lasso.shape[1], 1)
X_test_lasso_reshaped = X_test_lasso.reshape(X_test_lasso.shape[0], X_test_lasso.shape[1], 1)

### Convolutional Neural Network (CNN) Classifier

# Build the model
def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(3, activation='softmax'))  # 3 classes: Normal, Squamous, Adenocarcinoma
    
    model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the CNN model
cnn_model = build_cnn((X_train_lasso_reshaped.shape[1], 1))
history = cnn_model.fit(X_train_lasso_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2)

# Evaluate the model on test data
loss, accuracy = cnn_model.evaluate(X_test_lasso_reshaped, y_test)

# Save model metrics
model_metrics = {
    'Accuracy': accuracy,
    'Loss': loss,
    'Training Time (epochs)': history.epoch[-1]
}
model_metrics_df = pd.DataFrame(model_metrics, index=[0])
model_metrics_df.to_csv(os.path.join(output_folder, "model_metrics.csv"), index=False)

# Predict on the test set
y_pred = cnn_model.predict(X_test_lasso_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Save classification report
classification_report_dict = classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_, output_dict=True)
classification_report_df = pd.DataFrame(classification_report_dict).transpose()
classification_report_df.to_csv(os.path.join(output_folder, "classification_report.csv"))

# Save confusion matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
conf_matrix_df.to_csv(os.path.join(output_folder, "confusion_matrix.csv"))

# Plot Confusion Matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_folder, "confusion_matrix_plot.png"), dpi=300)
plt.close()

### Plotting Training and Validation Curves
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_folder, "training_validation_plot.png"), dpi=300)
plt.close()

# Calculate Precision, Recall, F1 Score
precision = classification_report_dict["weighted avg"]["precision"]
recall = classification_report_dict["weighted avg"]["recall"]
f1_score = classification_report_dict["weighted avg"]["f1-score"]

# Save additional metrics
additional_metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall (Sensitivity)": recall,
    "F1 Score": f1_score,
    "Training Time (epochs)": history.epoch[-1]
}
additional_metrics_df = pd.DataFrame(additional_metrics, index=[0])
additional_metrics_df.to_csv(os.path.join(output_folder, "additional_metrics.csv"), index=False)

print("All results and plots saved successfully in Lasso_CNN folder!")

# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "Lasso_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "Lasso_CNN_ROC_curve.png"), dpi=300)
plt.close()

```

```python
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import random
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve, auc, 
                             precision_recall_curve, f1_score, precision_score, recall_score)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from scipy.stats import entropy
import seaborn as sns
import matplotlib.pyplot as plt
import time
import csv

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Create output directory
output_dir = "KLDivergence_CNN"
os.makedirs(output_dir, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)
y = data['Condition']

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# KL Divergence for Feature Selection
def apply_kl_divergence(X_train, y_train, X_test, n_features=1000):
    kl_scores = []
    n_classes = y_train.shape[1]

    for col in X_train.T:
        kl_divergence_sum = 0
        for class_idx in range(n_classes):
            class_samples = col[y_train.argmax(axis=1) == class_idx]
            p_feature, _ = np.histogram(col, bins=10, density=True)
            p_class, _ = np.histogram(class_samples, bins=10, density=True)
            p_feature += 1e-8
            p_class += 1e-8
            kl_divergence_sum += entropy(p_class, p_feature)
        kl_scores.append(kl_divergence_sum)

    kl_indices = np.argsort(kl_scores)[::-1][:n_features]
    X_train_kl = X_train[:, kl_indices]
    X_test_kl = X_test[:, kl_indices]

    # Save selected features with scores
    selected_features = pd.DataFrame({
        "Gene": X.columns[kl_indices],
        "KL_Score": np.array(kl_scores)[kl_indices]
    })
    selected_features.to_csv(os.path.join(output_dir, "selected_features_kl.csv"), index=False)
    return X_train_kl, X_test_kl

# Apply KL Divergence
X_train_kl, X_test_kl = apply_kl_divergence(X_train_scaled, y_train, X_test_scaled)

# Reshape for CNN
X_train_kl_reshaped = X_train_kl.reshape(X_train_kl.shape[0], X_train_kl.shape[1], 1)
X_test_kl_reshaped = X_test_kl.reshape(X_test_kl.shape[0], X_test_kl.shape[1], 1)

# Build CNN model
def build_cnn(input_shape):
    model = Sequential([
        Conv1D(64, 3, activation='relu', input_shape=input_shape),
        MaxPooling1D(2),
        Dropout(0.3),
        Conv1D(32, 3, activation='relu'),
        MaxPooling1D(2),
        Dropout(0.3),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(3, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the CNN model
cnn_model = build_cnn((X_train_kl_reshaped.shape[1], 1))
start_time = time.time()
history = cnn_model.fit(X_train_kl_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2, callbacks=[early_stopping])
training_time = time.time() - start_time

# Save the trained model
cnn_model.save(os.path.join(output_dir, "cnn_model_kl.h5"))

# Evaluate the model on test data
loss, accuracy = cnn_model.evaluate(X_test_kl_reshaped, y_test)
print(f"Test Accuracy: {accuracy:.2f}")

# Predict on the test set
y_pred = cnn_model.predict(X_test_kl_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Compute classification metrics
f1 = f1_score(y_test_classes, y_pred_classes, average='weighted')
precision = precision_score(y_test_classes, y_pred_classes, average='weighted')
recall = recall_score(y_test_classes, y_pred_classes, average='weighted')
roc_auc = roc_auc_score(y_test, y_pred, average='weighted', multi_class='ovr')

# Save metrics to CSV
metrics = {
    "Test_Loss": loss,
    "Test_Accuracy": accuracy,
    "F1_Score": f1,
    "Precision": precision,
    "Recall": recall,
    "ROC_AUC": roc_auc,
    "Training_Time": training_time
}
with open(os.path.join(output_dir, "model_metrics.csv"), "w") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(metrics.keys())
    writer.writerow(metrics.values())

# Classification Report
class_report = classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_, output_dict=True)
pd.DataFrame(class_report).transpose().to_csv(os.path.join(output_dir, "classification_report.csv"))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.savefig(os.path.join(output_dir, "confusion_matrix.svg"), format='svg')

# ROC Curve
plt.figure()
fpr, tpr, _ = roc_curve(y_test.ravel(), y_pred.ravel())
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.savefig(os.path.join(output_dir, "roc_curve.svg"), format='svg')

# Precision-Recall Curve
plt.figure()
precision, recall, _ = precision_recall_curve(y_test.ravel(), y_pred.ravel())
plt.plot(recall, precision, color='purple', lw=2)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.savefig(os.path.join(output_dir, "precision_recall_curve.svg"), format='svg')

# Plotting Training and Validation Curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_dir, "training_validation_curves.svg"), format='svg')
plt.show()


# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "KLDivergence_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "Kldivergence_CNN_ROC_curve.png"), dpi=300)
plt.close()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import os
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import (classification_report, confusion_matrix, roc_curve, auc, 
                             precision_recall_curve, accuracy_score, precision_score, recall_score)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import seaborn as sns
import matplotlib.pyplot as plt

# Set seeds for reproducibility
np.random.seed(42)
import random
random.seed(42)
import tensorflow as tf
tf.random.set_seed(42)

# Create output directory
output_dir = "SelectFROM_CNN"
os.makedirs(output_dir, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)  # Drop non-feature columns
y = data['Condition']  # Target variable: Condition

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert class labels to numerical form (e.g., 0, 1, 2)
y_categorical = to_categorical(y_encoded)   # One-hot encode target for neural network

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Feature Selection using SelectFromModel
def select_features(X_train, y_train, X_test, n_features=1000):
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train.argmax(axis=1))  # Use original class labels for training

    selector = SelectFromModel(rf_model, max_features=n_features, prefit=True)
    X_train_selected = selector.transform(X_train)
    X_test_selected = selector.transform(X_test)
    
    selected_genes = X.columns[selector.get_support(indices=True)]
    selected_scores = rf_model.feature_importances_[selector.get_support()]
    selected_genes_df = pd.DataFrame({"Gene": selected_genes, "Score": selected_scores})
    selected_genes_df.to_csv(f"{output_dir}/selected_genes.csv", index=False)
    
    return X_train_selected, X_test_selected, selector.get_support(indices=True)

# Apply SelectFromModel
X_train_selected, X_test_selected, selected_features = select_features(X_train_scaled, y_train, X_test_scaled, n_features=1000)

# Reshape the data for CNN
X_train_reshaped = X_train_selected.reshape(X_train_selected.shape[0], X_train_selected.shape[1], 1)
X_test_reshaped = X_test_selected.reshape(X_test_selected.shape[0], X_test_selected.shape[1], 1)

# Build the model
def build_cnn(input_shape):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.4))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.4))
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(len(np.unique(y)), activation='softmax'))  # Adjust for number of classes
    
    model.compile(optimizer=Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the CNN model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

cnn_model = build_cnn((X_train_reshaped.shape[1], 1))

# Start training timer
start_time = time.time()

# Fit the model on training data
history = cnn_model.fit(
    X_train_reshaped, y_train, 
    epochs=50, batch_size=16, 
    validation_split=0.2, 
    callbacks=[early_stopping]
)

# End training timer
training_time = time.time() - start_time

# Evaluate the model on test data
loss, accuracy = cnn_model.evaluate(X_test_reshaped, y_test)
print(f"Test Accuracy: {accuracy:.2f}")

# Model Evaluation Metrics
y_pred = cnn_model.predict(X_test_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Save metrics
classification_metrics = classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_, output_dict=True)
metrics_df = pd.DataFrame(classification_metrics).transpose()
metrics_df.loc["Training Time"] = training_time
metrics_df.to_csv(f"{output_dir}/metrics.csv", index=True)

# Calculate summary metrics for additional CSV
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes, average='weighted')
recall = recall_score(y_test_classes, y_pred_classes, average='weighted')
f1 = (2 * (precision * recall)) / (precision + recall) if (precision + recall) > 0 else 0

# Create a DataFrame for the additional metrics
additional_metrics = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'Training Time'],
    'Value': [accuracy, precision, recall, f1, training_time]
})

# Save the additional metrics to a new CSV file
additional_metrics.to_csv(f"{output_dir}/additional_metrics.csv", index=False)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.savefig(f"{output_dir}/confusion_matrix.png", dpi=300)
plt.close()

### Plotting Training and Validation Curves

# Plot training and validation accuracy and loss
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(f"{output_dir}/training_validation_curves.png", dpi=300)
plt.close()

### ROC Curve and Precision-Recall Curve

# ROC Curve for each class
from sklearn.metrics import roc_auc_score

# Binarize the output for ROC curve
y_test_bin = y_test
y_pred_bin = y_pred

plt.figure()
for i in range(len(label_encoder.classes_)):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_bin[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'ROC curve for {label_encoder.classes_[i]} (area = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.savefig(f"{output_dir}/roc_curve.png", dpi=300)
plt.close()

# Precision-Recall Curve for each class
plt.figure()
for i in range(len(label_encoder.classes_)):
    precision, recall, _ = precision_recall_curve(y_test_bin[:, i], y_pred_bin[:, i])
    plt.plot(recall, precision, marker='.', label=f'Precision-Recall curve for {label_encoder.classes_[i]}')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend(loc="lower left")
plt.savefig(f"{output_dir}/precision_recall_curve.png", dpi=300)
plt.close()

print("All Results are stored in the folder")


# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "SelectFromModel_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "SelectFromModel_CNN_ROC_curve.png"), dpi=300)
plt.close()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import os
import random
import tensorflow as tf
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, Activation
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import label_binarize
from tensorflow.keras.callbacks import EarlyStopping
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, precision_recall_curve

# Set seeds for reproducibility
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Create directory to store results
output_dir = 'HYBRIDMODEL'
os.makedirs(output_dir, exist_ok=True)

# Load Data
data = pd.read_csv('TCGA_ICGC_Commond_Genes_Merged_Mean-Imputed_SMOTE.csv')  # Replace with your file path

# Preprocessing
X = data.drop(['Sample_ID', 'Condition'], axis=1)  # Drop non-feature columns
y = data['Condition']  # Target variable: Condition

# Encoding labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert class labels to numerical form (e.g., 0, 1, 2)
y_categorical = to_categorical(y_encoded)   # One-hot encode target for neural network

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.3, random_state=42)

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape the data for CNN
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)  # (samples, features, 1)
X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

### Feature Extraction

# Method 1: PCA for Dimensionality Reduction
def apply_pca(X_train, X_test, n_components=1000):  # Increase to 500 or 1000
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    print(f"Explained variance by PCA: {sum(pca.explained_variance_ratio_):.2f}")
    return X_train_pca, X_test_pca

# Method 2: Mutual Information for Feature Selection
def apply_mutual_info(X_train, X_test, y_train, num_features=500):  # Increase to 500 or 1000
    mi_selector = SelectKBest(mutual_info_classif, k=num_features)
    X_train_mi = mi_selector.fit_transform(X_train, np.argmax(y_train, axis=1))  # Select top features based on Mutual Information
    X_test_mi = mi_selector.transform(X_test)
    selected_features = mi_selector.get_support(indices=True)
    selected_feature_names = X.columns[selected_features]  # Get the names of selected features
    print(f"Selected top {num_features} features based on Mutual Information")

    # Save selected important features to a CSV file
    important_features_df = pd.DataFrame({'Important_Features': selected_feature_names, 
                                           'MI_Score': mi_selector.scores_[selected_features]})
    important_features_df.to_csv(os.path.join(output_dir, 'important_features.csv'), index=False)
    print(f"Important features saved to '{os.path.join(output_dir, 'important_features.csv')}'")
    
    return X_train_mi, X_test_mi

# Hybrid Feature Selection (PCA + Mutual Information)
X_train_pca, X_test_pca = apply_pca(X_train_scaled, X_test_scaled, n_components=1000)  # Increase to 500 or 1000
X_train_hybrid, X_test_hybrid = apply_mutual_info(X_train_pca, X_test_pca, y_train, num_features=500)  # Increase to 500 or 1000

# Reshape the data for CNN
X_train_hybrid_reshaped = X_train_hybrid.reshape(X_train_hybrid.shape[0], X_train_hybrid.shape[1], 1)
X_test_hybrid_reshaped = X_test_hybrid.reshape(X_test_hybrid.shape[0], X_test_hybrid.shape[1], 1)

### Convolutional Neural Network (CNN) Classifier

# Build the model with ReLU activation
def build_cnn(input_shape):
    model = Sequential()
    
    model.add(Conv1D(filters=64, kernel_size=4, input_shape=input_shape))
    model.add(Activation('relu'))  # Use ReLU activation instead of LeakyReLU
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    
    model.add(Conv1D(filters=32, kernel_size=3))
    model.add(Activation('relu'))  # Use ReLU activation instead of LeakyReLU
    model.add(MaxPooling1D(pool_size=2))
    model.add(Dropout(0.5))
    
    model.add(Flatten())
    model.add(Dense(64))
    model.add(Activation('relu'))  # Use ReLU activation instead of LeakyReLU
    model.add(Dropout(0.5))
    
    model.add(Dense(3, activation='softmax'))  # Output layer for 3 classes
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Train the CNN model
cnn_model = build_cnn((X_train_hybrid_reshaped.shape[1], 1))

# Early stopping to stop training when validation loss doesn't improve after 3 epochs
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Fit the model on training data and record training time
import time
start_time = time.time()
history = cnn_model.fit(X_train_hybrid_reshaped, y_train, epochs=50, batch_size=16, validation_split=0.2, callbacks=[early_stopping])
training_time = time.time() - start_time

# Evaluate the model on test data
loss, accuracy = cnn_model.evaluate(X_test_hybrid_reshaped, y_test)
print(f"Test Accuracy: {accuracy:.2f}")

# Predict on the test set
y_pred = cnn_model.predict(X_test_hybrid_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Classification Report
precision, recall, f1, _ = precision_recall_fscore_support(y_test_classes, y_pred_classes, average='weighted')
print("Classification Report:\n", classification_report(y_test_classes, y_pred_classes, target_names=label_encoder.classes_))

# Save metrics to a CSV file
metrics_df = pd.DataFrame({
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall': [recall],
    'F1_Score': [f1],
    'Training_Time': [training_time]
})
metrics_df.to_csv(os.path.join(output_dir, 'model_metrics.csv'), index=False)
print(f"Model metrics saved to '{os.path.join(output_dir, 'model_metrics.csv')}'")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.savefig(os.path.join(output_dir, 'confusion_matrix.png'), dpi=300)
plt.show()

### Plotting Training and Validation Curves

# Plotting the training and validation accuracy and loss
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'accuracy_loss_curves.png'), dpi=300)
plt.show()

### ROC Curve
# Assuming binary classification for the second class (you may adjust the index based on your class of interest)
fpr, tpr, thresholds = roc_curve(y_test_classes, y_pred[:, 1], pos_label=1)  # Adjust pos_label for your case
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='blue', label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line for random guessing
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.savefig(os.path.join(output_dir, 'roc_curve.png'), dpi=300)
plt.show()


# for auc-roc

# Calculate overall AUC-ROC for multi-class classification
roc_auc = roc_auc_score(y_test, y_pred, multi_class='ovr')

# Save the single AUC-ROC value
with open(os.path.join(output_folder, "Hrbrid_CNN_auc_roc_value.txt"), "w") as file:
    file.write(f"Overall AUC-ROC: {roc_auc:.2f}\n")

# Plot and save ROC Curve for the overall classification
plt.figure(figsize=(8, 6))
for i, class_name in enumerate(label_encoder.classes_):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'{class_name}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC = {roc_auc:.2f})")
plt.legend()
plt.savefig(os.path.join(output_folder, "Hybrid_CNN_ROC_curve.png"), dpi=300)
plt.close()
```
